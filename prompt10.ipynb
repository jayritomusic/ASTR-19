```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
random_numbers = np.random.rand(1000)
```

```python
plt.figure(figsize=(10, 5))
plt.hist(random_numbers, bins=100, color='skyblue', edgecolor='black', alpha=0.7)

plt.title('Histogram of 1000 Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.savefig('random_numbers_histogram.pdf')

plt.show()
```
