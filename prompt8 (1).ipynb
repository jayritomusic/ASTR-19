```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create an array x ranging from 0 to 1 in 100 steps
x = np.linspace(0, 1, 100)
```

```python
# Define the function for sin(x)
def sin_function(x):
    return np.sin(2 * np.pi * x)  # Scale to fit one full sine wave in the range [0, 1]

# Define the function for cos(x)
def cos_function(x):
    return np.cos(2 * np.pi * x)  # Scale to fit one full cosine wave in the range [0, 1]
```

```python
# Create a multipanel plot (1 row, 2 columns)
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Plot sin(x) vs. x in the left panel
axs[0].plot(x, sin_function(x), color='b')
axs[0].set_title('sin(x)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('sin(x)')
axs[0].grid()

# Plot cos(x) vs. x in the right panel
axs[1].plot(x, cos_function(x), color='r')
axs[1].set_title('cos(x)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('cos(x)')
axs[1].grid()

# Adjust layout
plt.tight_layout()

# Save the figure as a PDF
plt.savefig('trig_functions.pdf')

# Show the plot
plt.show()
```
