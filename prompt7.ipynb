```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)
```

```python
def exp_function(x):
    return np.exp(x)
```

```python
y = exp_function(x)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = exp(x)', color='blue')
plt.title('Exponential Function')
plt.xlabel('Time [milliseconds]')
plt.ylabel('Awesomeness')
plt.legend()
plt.grid()
```
